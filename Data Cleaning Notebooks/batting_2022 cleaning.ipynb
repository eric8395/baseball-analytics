```python
from pybaseball import batting_stats
import numpy as np
import pandas as pd


# get all batting data 
batting_2022 = batting_stats(2022, ind = 1, qual = 50)
```

```python
batting_2022.head()
```

```python
pd.set_option('display.max_rows', 10)

pd.set_option('display.max_columns', 500)
batting_2022 = batting_2022[['Name', 'Season', 'Age', 'H', 'R', 'RBI', 'BB', 'SO', 'SB', 'OPS']]
```

```python
batting_2022.head()
```

```python
from pybaseball import bwar_bat

# get war stats from baseball reference 
bwar_bat = bwar_bat()

# remove excess stats; just need salary
cols_to_remove = ['runs_above_avg','runs_above_avg_off', 'runs_above_avg_def', 
                  'WAR_rep', 'WAA', 'WAR']

bwar_bat = bwar_bat.drop(columns = cols_to_remove)
bwar_bat.head()
```

```python
# filter bwar_bat 2022 and minimum 100 PA
bwar_bat_2022 = bwar_bat[(bwar_bat['year_ID'].between(2005, 2022)) & (bwar_bat['PA'] >= 100)]
bwar_bat_2022.sort_values(by = 'name_common')
```

```python
bwar_bat_2022['Avg Career Salary Difference'] = bwar_bat_2022['salary'].diff()
bwar_bat_2022 = bwar_bat_2022.fillna(0)
bwar_bat_2022
```

```python
# get 2022 salaries only
salary_2022 = bwar_bat_2022[['name_common','salary', 'year_ID']]
salary_2022 = salary_2022[salary_2022['year_ID'] == 2022]
salary_2022
```

```python
# group by Name and find averages across all other features
bwar_bat_2022 = bwar_bat_2022.groupby(['name_common']).mean()
bwar_bat_2022 = bwar_bat_2022.drop(columns = ['mlb_ID', 'year_ID', 'stint_ID', 'G', 'PA'])
bwar_bat_2022 = bwar_bat_2022.reset_index()
```

```python
bwar_bat_2022.head()
```

```python
# join batting_2022 onto bwar_bat_2022 (left join)
batting_merged = pd.merge(bwar_bat_2022, 
                          batting_2022, 
                          how='left',
                          left_on=['name_common'],right_on=['Name'])
```

```python
batting_merged[batting_merged['name_common'] == 'Mike Zunino']
```

```python
batting_merged = batting_merged.dropna()
batting_merged
```

```python
batting_merged = batting_merged.drop(columns = ['salary', 'name_common'])
```

```python
batting_merged
```

```python
salary_2022 = salary_2022.rename(columns = {'salary':'2022 Salary'})
salary_2022
```

```python
# join 2022 salaries
merged_salaries = pd.merge(salary_2022,
                           batting_merged, 
                           how='left',
                           left_on=['name_common'],right_on=['Name'])
```

```python
merged_salaries
```

```python
merged_salaries.columns
```

```python
batting_merged_2022 = merged_salaries[['Name', '2022 Salary', 'Avg Career Salary Difference', 'Age', 'H', 'R', 'RBI', 'BB', 'SO', 'SB', 'OPS' ]]
```

```python
batting_merged_2022 = batting_merged_2022.dropna()
```

```python
batting_merged_2022
```

```python
batting_merged_2022['2022 Salary'] = batting_merged_2022['2022 Salary'].replace(0, 700000)
```
