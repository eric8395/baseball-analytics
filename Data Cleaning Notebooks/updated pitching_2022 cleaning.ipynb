```python
from pybaseball import pitching_stats
import numpy as np
import pandas as pd


# get all pitching data 
pitching_2022 = pitching_stats(2022, ind = 1, qual = 30)
```

```python
pitching_2022.head()
```

```python
pitching_2022.columns
```

```python
pd.set_option('display.max_rows', 10)

pd.set_option('display.max_columns', 500)
pitching_2022 = pitching_2022[['Name', 'Season', 'Age', 'W', 'L', 'ERA', 'G', 'SV', 'IP', 'H', 'HR', 'SO', 'BB']]
```

```python
pitching_2022.head()
```

```python
from pybaseball import bwar_pitch

# get war stats from baseball reference 
bwar_pitch = bwar_pitch(return_all = True)

# remove excess stats; just need salary
cols_to_keep = ['name_common', 'age', 'player_ID', 'year_ID', 'salary']

bwar_pitch = bwar_pitch[cols_to_keep]
bwar_pitch.head()
```

```python
# filter bwar_pitch_2022
bwar_pitch_2022 = bwar_pitch[(bwar_pitch['year_ID'].between(2005, 2022))]
bwar_pitch_2022.sort_values(by = 'name_common')
```

```python
bwar_pitch_2022['Avg Career Salary Difference'] = bwar_pitch_2022['salary'].diff()
bwar_pitch_2022 = bwar_pitch_2022.fillna(0)
bwar_pitch_2022
```

```python
# get 2022 salaries only
salary_2022 = bwar_pitch_2022[['name_common','salary', 'year_ID']]
salary_2022 = salary_2022[salary_2022['year_ID'] == 2022]
salary_2022
```

```python
# group by Name and find averages across all other features
bwar_pitch_2022 = bwar_pitch_2022.groupby(['name_common']).mean()
bwar_pitch_2022
```

```python
bwar_pitch_2022 = bwar_pitch_2022.drop(columns = ['year_ID'])
bwar_pitch_2022 = bwar_pitch_2022.reset_index()
```

```python
bwar_pitch_2022.head()
```

```python
# join pitching_2022 onto bwar_pitch_2022 (left join)
pitching_merged = pd.merge(bwar_pitch_2022, 
                          pitching_2022, 
                          how='left',
                          left_on=['name_common'],right_on=['Name'])
```

```python
pitching_merged[pitching_merged['name_common'] == 'Gerrit Cole']
```

```python
pitching_merged = pitching_merged.dropna()
pitching_merged
```

```python
pitching_merged = pitching_merged.drop(columns = ['salary', 'name_common', 'age'])
```

```python
pitching_merged
```

```python
salary_2022 = salary_2022.rename(columns = {'salary':'2022 Salary'})
salary_2022
```

```python
# join 2022 salaries
merged_salaries = pd.merge(salary_2022,
                           pitching_merged, 
                           how='left',
                           left_on=['name_common'],right_on=['Name'])
```

```python
merged_salaries
```

```python
merged_salaries.columns
```

```python
pitching_merged_2022 = merged_salaries[['Name', '2022 Salary', 'Avg Career Salary Difference', 
                                        'Age', 'W', 'L', 'ERA', 'G', 'SV', 'IP', 'H', 'HR', 'SO', 'BB' ]]
```

```python
updated_pitching_merged_2022 = pitching_merged_2022.dropna()
```

```python
updated_pitching_merged_2022
```

```python
updated_pitching_merged_2022['2022 Salary'] = updated_pitching_merged_2022['2022 Salary'].replace(0, 700000)
```
