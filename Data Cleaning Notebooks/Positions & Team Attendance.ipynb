```python
import pandas as pd
from pybaseball.lahman import *
pd.set_option('display.max_columns', None)
```

```python
fielding = fielding()
fielding
```

```python
# get name, year and POS
positions = fielding[['playerID', 'yearID', 'POS']]

# filter to just 2000 - 2021
positions = positions[positions['yearID'] >= 2000]
positions
```

```python
positions['POS'].value_counts()
```

```python
positions.shape
```

```python
positions['POS'].isna().sum()
```

```python
path = "Data/"

positions.to_csv(path + 'positions')
```

## Get Team Attendance (1900 - 2021)

```python
# data on teams by year: record, division, stadium, attendance, etc
teams = teams()
```

```python
teams.tail()
```

```python
attendance = teams[['yearID', 'name', 'franchID', 'attendance', 'W', 'L']]
attendance
```

```python
attendance = attendance[attendance['yearID'] >= 1900]
```

```python
attendance = attendance.dropna()
```

```python
attendance.isna().sum()
```

```python
attendance
```

```python
path = "Data/"

attendance.to_csv(path + 'attendance')
```
