## Collect Team Statistics since 1960

Below we will import team batting, pitching, and fielding statistics since 1960. 

1960 will be used as a benchmark as this was approximately the time when baseball started its 'Expansion Era'. In other words, baseball implemented new rules close to what we see to this day and expanded to 30 teams. 

**Team Batting**

```python
from pybaseball import team_batting

# get each team's seasonal batting stats from 1960 through 2021
team_batting = team_batting(1960, 2021)
```

```python
team_batting.sort_values(by = 'Season', ascending = False)
```

```python
pd.set_option('display.max_rows', None) # change to None show all columns 
team_batting.isna().sum()
```

**Team Pitching**

```python
from pybaseball import team_pitching

# get each team's individual-season-level pitching stats from 1960 through 2021
team_pitching = team_pitching(1960, 2021, ind = 1)
```

```python
team_pitching.sort_values(by = 'Season', ascending = False)
```

```python
team_pitching[team_pitching['Team'] == 'NYY'].sort_values(by = 'Season').head()
```

```python
pd.set_option('display.max_rows', None) # change to None show all columns 
team_pitching.isna().sum()
```

**Team Fielding**

Note: Team Fielding table will be limited to data between 2011-2021. There is a scraping error whereby data before the StatCast era will not be brought in, and will result in incorrect table that does not reflect fielding data. 

For the purposes of this analysis, we will strictly limit the analysis to team fielding statistics from 2011-2021. 

```python
from pybaseball import team_fielding

# get each team's seasonal fielding stats from 2011 through 2021
team_fielding = team_fielding(2011, 2021)
```

```python
team_fielding.sort_values(by = 'Season', ascending = False).head()
```

```python
pd.set_option('display.max_rows', None) # change to None show all columns 
team_fielding.isna().sum()
```
